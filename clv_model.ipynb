```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 5000

data = pd.DataFrame({
    "CustomerID": np.random.randint(1000, 2000, n),
    "PurchaseDate": pd.date_range(start="2023-01-01", periods=n, freq="H"),
    "OrderValue": np.random.randint(100, 5000, n)
})

data.to_csv("ecommerce_dummy.csv", index=False)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("ecommerce_dummy.csv")
df["PurchaseDate"] = pd.to_datetime(df["PurchaseDate"])

summary = df.groupby("CustomerID").agg({
    "OrderValue": ["sum", "count", "mean"]
})

summary.columns = ["TotalRevenue", "Frequency", "AvgOrderValue"]

summary["CLV"] = summary["TotalRevenue"] * 1.2  # simple projected multiplier

summary = summary.sort_values("CLV", ascending=False)

print(summary.head())

#CLV = AvgOrderValue × Purchase Frequency × Retention Time
```
